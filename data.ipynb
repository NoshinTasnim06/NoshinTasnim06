```python
df_bb100 = pd.read_csv('/Users/noshintasnim/Downloads/hot-100-current.csv.zip')
df_classics = pd.read_csv('/Users/noshintasnim/Downloads/ClassicHit.csv')
```

```python
df_bb100_top = df_bb100.loc[
    df_bb100.groupby(['title', 'performer'])['wks_on_chart'].idxmax()
].reset_index(drop=True)

df_bb100_top.head(2)
```

```python
best_songs = pd.merge(
    df_bb100_top, 
    df_classics, 
    left_on= ['title', 'performer'], 
    right_on= ['Track', 'Artist'], 
    how='inner' 
)

best_songs = best_songs.sort_values('Year').reset_index(drop=True)
best_songs.shape
```

```python
best_songs.head(2)
```

```python
best_songs.duplicated().sum()
```

```python
best_songs.info()
best_songs.describe(include='all').transpose()
best_songs.isnull().sum().sort_values(ascending=False)
```

```python
cols_to_drop = ['Track', 'performer', 'last_week', 'Time_Signature', 'BPM', 'Key',
                'Mode', 'current_week', 'Valence', 'chart_week', 'Acousticness', 'Instrumentalness']

dfs = [best_songs]

for df in dfs:
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_' )
    df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)

expected_columns = [
    'year', 'title', 'artist', 'genre', 'peak_pos', 'wks_on_chart', 
    'duration', 'popularity', 'energy', 'danceability', 'loudness', 
    'speechiness', 'liveness']

existing_cols = [col for col in expected_columns if col in best_songs.columns]
best_songs = best_songs[existing_cols]

```
